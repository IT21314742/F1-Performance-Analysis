## Extract Qualifying and Race Results

```python
# Cell 4: Create grid vs finish comparison
def get_grid_vs_finish(qualy_session, race_session):
    """
    Extract qualifying positions (grid) and race finishing positions
    """
    # Get qualifying results
    qualy_results = qualy_session.results[['Abbreviation', 'Position', 'TeamName']].copy()
    qualy_results.rename(columns={'Position': 'Grid_Position', 'Abbreviation': 'Driver'}, inplace=True)
    
    # Get race results
    race_results = race_session.results[['Abbreviation', 'Position', 'Status']].copy()
    race_results.rename(columns={'Position': 'Race_Position', 'Abbreviation': 'Driver'}, inplace=True)
    
    # Merge the data
    comparison = pd.merge(qualy_results, race_results, on='Driver', how='outer')
    
    # Calculate positions gained/lost
    comparison['Grid_Position'] = pd.to_numeric(comparison['Grid_Position'], errors='coerce')
    comparison['Race_Position'] = pd.to_numeric(comparison['Race_Position'], errors='coerce')
    comparison['Positions_Changed'] = comparison['Grid_Position'] - comparison['Race_Position']
    
    # Sort by grid position
    comparison = comparison.sort_values('Grid_Position').reset_index(drop=True)
    
    return comparison

# Get comparison for Monaco 2023
monaco_comparison = get_grid_vs_finish(qualy, race)
print("🏁 Monaco 2023: Grid vs Race Finish")
print(monaco_comparison[['Driver', 'TeamName', 'Grid_Position', 'Race_Position', 'Positions_Changed']])

# Find biggest movers
best_gain = monaco_comparison.loc[monaco_comparison['Positions_Changed'].idxmax()]
most_lost = monaco_comparison.loc[monaco_comparison['Positions_Changed'].idxmin()]

print(f"\n📈 Biggest gainer: {best_gain['Driver']} (+{best_gain['Positions_Changed']} positions)")
print(f"📉 Biggest loser: {most_lost['Driver']} ({most_lost['Positions_Changed']} positions)")
```

```python
# Visualize Grid vs Finish 
def plot_grid_vs_finish(comparison, race_name):
    """
    Create a visual comparing grid and finishing positions
    """
    fig, ax = plt.subplots(figsize=(12, 8))
    
    
    # Set up positions for plotting
    y_pos = range(len(comparison))
    width = 0.35
    
    #plot bars
    grid_bars = ax.barh([y - width/2 for y in y_pos],
                        comparison['Grid_Position'],
                        width, label='Grid_Position', color='lightgray')
    race_bars = ax.barh([y + width/2 for y in y_pos],
                        comparison['Race_Position'],
                        width, label='Race Finish', color='crimson')
    
    # Customize Plot
    ax.set_yticks(y_pos)
    ax.set_yticklabels([f"{row['Driver']} ({row['TeamName']})" for _, row in comparison.iterrows()])
    ax.set_xlabel('Position')
    ax.set_title(f'{race_name}: Grid Position vs Race Finish', fontsize=16, fontweight='bold')
    ax.legend()
    ax.invert_xaxis() #Position 1 should be at the top
    ax.invert_yaxis() # Grid Position 1 at top
    
    
    #Add grid for readability
    ax.grid(True, alpha=0.3, axis='x')
    
    
```
